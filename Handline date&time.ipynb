## Working with dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## Extract year

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

## Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

## Extract days 

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

## Extract week of the year 

```python
date['date_week'] = date['date'].dt.isocalendar().week

# Drop columns and display head
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

## Extract quarter 

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

## extract time elapsed between dates 

```python
#"Extract time elapsed between dates" means calculating the difference between two dates to find out how much time has passed between them.
```

```python
import datetime
## here we can find out the present date or time 
today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

## Working on time

```python
time.info()
```

```python
##You need to convert the 'date' column to datetime format first using pd.to_datetime():
time['date'] = pd.to_datetime(time['date'])  # Convert to datetime
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

## Extract time part

```python
time['time'] = time['date'].dt.time

time.head()
```

## Time difference

```python
today - time['date']
```

```python
# in seconds
# calculates the number of seconds that have passed between each datetime in time['date'] and today.

(today - time['date'])/np.timedelta64(1,'s')
```
